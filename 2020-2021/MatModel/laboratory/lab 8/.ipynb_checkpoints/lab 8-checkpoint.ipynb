```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
```

## Начальные значения

```python
t0 = 0
t = np.arange(t0, 20, 0.01)

x0 = [7, 7.7]
p_cr = 27
N = 37
q = 1

tau1 = 17
tau2 = 16

p1 = 15
p2 = 12

d = 0.00024
```

```python
a1 = p_cr / (tau1*tau1 * p1*p1 * N * q)
a2 = p_cr / (tau2*tau2 * p2*p2 * N * q)
b = p_cr / (tau1*tau1 * p1*p1 * tau2*tau2 * p2*p2 * N * q)
c1 = (p_cr - p1) / (tau1 * p1)
c2 = (p_cr - p2) / (tau2 * p2)
```

## Система уравнения первого случая

```python
def syst1 (y, t):
    y1, y2 = y
    return [y1 - (b/c1)*y1*y2 -(a1/c1)*y1*y1, (c2/c1)*y2 - (b/c1)*y1*y2 - (a2/c1)*y2*y2]
```

## Система уравнения второго случая

```python
def syst2 (y, t):
    y1, y2 = y
    return [y1 - (b/c1)*y1*y2 -(a1/c1)*y1*y1, (c2/c1)*y2 - ((b/c1) + d)*y1*y2 - (a2/c1)*y2*y2]
```

```python
y1 = odeint (syst1, x0, t)
y2 = odeint (syst2, x0, t)

M11 = y1[:,0]
M12 = y1[:,1]

M21 = y2[:,0]
M22 = y2[:,1]
```

## Первый случай

```python
fig = plt.figure (facecolor = 'white')
plt.plot (t, M11, linewidth = 2, label = 'M1')
plt.plot (t, M12, linewidth = 2, label = 'M2')
plt.xlabel ("Время")
plt.ylabel ("Капитал")
plt.grid (True)
plt.legend ()
plt.show ()
fig.savefig ('image/01.png', dpi = 600)
```

## Второй случай

```python
fig2 = plt.figure (facecolor = 'white')
plt.plot (t, M21, linewidth = 2, label = 'M1')
plt.plot (t, M22, linewidth = 2, label = 'M2')
plt.xlabel ("Время")
plt.ylabel ("Капитал")
plt.grid (True)
plt.legend ()
plt.show ()
fig2.savefig ('image/02.png', dpi = 600)
```
